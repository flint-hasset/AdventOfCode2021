### Part 1

```python
hex_bin = {'0':'0000','1':'0001','2':'0010','3':'0011','4':'0100','5':'0101','6':'0110','7':'0111','8':'1000','9':'1001','A':'1010','B':'1011','C':'1100','D':'1101','E':'1110','F':'1111'}
bin_dec = {'000':0, '001':1, '010':2, '011':3, '100':4, '101':5, '110':6, '111':7}
#bin_hex = {v: k for k, v in hex_bin.items()}
```

```python
def check_type(bin_str):
    version = bin_dec[bin_str[:3]]
    ptype = bin_dec[bin_str[3:6]]
    
    if ptype == 4:
        print('Literal, use lit rules')
        ptype  = 'Literal'
    else:
        print('Operator, use op rules')
        ptype = 'Operator'
    return version, ptype
    
def run_literal(bin_str):
    version = bin_dec[bin_str[:3]]
    rem_str = bin_str[6:]
    val_str = ''
    keep_going = 1
    packets = 0
    #print(rem_str)
    while keep_going == 1:
        packets +=1
        if rem_str[0] == '0':
            val_str = val_str+rem_str[1:5]
            keep_going = 0
            rem_str = rem_str[5:]
        elif rem_str[0] == '1':
            val_str = val_str+rem_str[1:5]
            rem_str = rem_str[5:]
    bits_used = len(bin_str)-len(rem_str)
    
    return int(val_str,2), rem_str, version, bits_used

def run_operator(bin_str):
    version = bin_str[3:6]
    
    ### Number of packets
    if bin_str[6] == '1':
        print('Packets')
        length = 11
        subpackets = int(bin_str[7:7+length],2)
        rem_substr = bin_str[7+length:]    
        num_packets_flag = 1
    else:
        ### Number of bits
        print("Bits")
        length = 15
        subpackets = bin_str[7:7+length]
        subpackets = int(subpackets,2)
        rem_substr = bin_str[7+length:]
        num_packets_flag = 0
    bits_used = len(bin_str)-len(rem_substr)
    
    return rem_substr, num_packets_flag, subpackets, version, bits_used
```

```python
def recursive_packet(bin_str):
    total_version = 0
    version, ptype = check_type(bin_str)
    total_version += version
    if ptype == 'Operator':
        rem_str, num_packets_flag, subpackets, version, bits_used = run_operator(bin_str)
#        total_version += version
        print("Add me, operator: ", total_version)

        if num_packets_flag == 1:
            for x in range(subpackets):
                ### Recursion here? 
                rem_str = recursive_packet(rem_str)

        else: 
            counter = 0
            while counter < subpackets:
                ### Not sure how to get bits used in w recursion. Not solved. 
                rem_str = recursive_packet(rem_str)
                counter += bits_used
                #print(version, counter)
    ### For literals
    else:
        val, rem_str, version, bits_used = run_literal(bin_str)
#        total_version += version
        print("Add me, literal: ", total_version)
    
    return rem_str
```
