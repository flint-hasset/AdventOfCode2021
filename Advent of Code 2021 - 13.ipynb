### Part 1

```python
input_set = a
```

```python
coordinates = []
folds = []
for x in input_set:
    if x=='':
        pass
    elif x[:4]=='fold':
        folds.append(x[11:])
    else:
        coordinates.append([int(x.split(',')[0]),int(x.split(',')[1])])
```

```python
max_x, max_y = max(x[0] for x in coordinates), max(x[1] for x in coordinates)
grid = [[0 for x in range(max_x+1)] for x in range(max_y+1)]

for cor in coordinates:
    grid[cor[1]][cor[0]] = 1

for fold in folds[:1]:

    if fold[0]=='y':
        spot = int(fold.split('=')[1])
        sec2 = grid[spot+1:][::-1]
        grid = grid[:spot]

    if fold[0]=='x':
        spot = int(fold.split('=')[1])
        sec2 = [x[spot+1:] for x in grid]
        sec2 = [x[::-1] for x in sec2]
        grid = [x[:spot] for x in grid]


    for y in range(len(grid)):
        for x in range(len(grid[0])):
            grid[y][x] = max(grid[y][x], sec2[y][x])

sum([sum(x) for x in grid])
```
